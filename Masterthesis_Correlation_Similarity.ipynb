```python
import pandas as pd
import pickle
import numpy as np
with open('closest_list.pickle', 'rb') as handle:
    closest_list = pickle.load(handle)
```

```python
from tqdm import tqdm

state = {
    "downloaded": 0,
    "adj_close_list": [],
    "raw_data_list": []
}

try:
    with open('state.pickle', 'rb') as handle:
        state = pickle.load(handle)
    print("loaded from file")
    print(f"already downloaded: {state['downloaded']} ({state['downloaded']/len(closest_list)}%)")
    print("continuing download")
except:
    print("starting download from the beginning")



import yfinance as yf

adj_close_list = []
start_from = state['downloaded']
counter = 0
for elem in tqdm(closest_list[start_from:]):
    print(f"saving {start_from + counter}")
    if len(elem) == 0:
        state['downloaded'] += 1
        state['adj_close_list'].append(pd.DataFrame())
        state['raw_data_list'].append(pd.DataFrame())
        counter += 1
        continue
    ticker_list = elem
    data = yf.download(ticker_list, start= '2020-1-1', end= '2023-1-1')
    adj_close= data['Adj Close']
    state['downloaded'] += 1
    state['adj_close_list'].append(adj_close)
    state['raw_data_list'].append(data)
    counter += 1
    

    with open('state.pickle', 'wb') as handle:
        pickle.dump(state, handle, protocol=pickle.HIGHEST_PROTOCOL)
        
print("finished downloading everything")
```

```python
from tqdm import tqdm


sec_returns_list = []
for adj_close in tqdm(adj_close_list):
    #adj_close = np.array(adj_close)
    sec_returns = np.log(adj_close/ adj_close.shift(1))
    sec_returns_list.append(sec_returns)
#sec_returns
```

```python
len(sec_returns_list)
```

```python
corr_matrix_list = []
for sec_returns in tqdm(sec_returns_list):
    sec_returns = pd.DataFrame(sec_returns)
    corr_matrix = sec_returns.corr()
    corr_matrix_list.append(corr_matrix)
```

```python
np.array(corr_matrix_list[15])[2][1]
```

```python
def matrix_to_triangle_value_array(matrix):
    coeff_list = []
    matrix = np.array(matrix)
    length = matrix.shape[0]
    for i in range(1, length):
        for j in range(i):
            value = matrix[i][j]
            if np.isnan(value):
                continue
            coeff_list.append(value)
    return coeff_list
```

```python
all_coeffs = []
for corr_matrix in tqdm(corr_matrix_list):
    all_coeffs += matrix_to_triangle_value_array(corr_matrix)
len(all_coeffs)
```

```python
import matplotlib.pyplot as plt
import numpy as np
result = np.array(all_coeffs)
num_bins = 100 # <-- Change here - Specify total number of bins for histogram
plt.hist(result.ravel(), bins=np.linspace(np.min(result), np.max(result), num=num_bins)) #<-- Change here.  Note the use of ravel.
plt.title('Frequency Distribution (Total Correlation Coefficients, N=822591)')
plt.xlabel('Correlation')
plt.ylabel('Frequency')
plt.savefig('Frequency Distribution (Total Correlation Coefficients, N=822591).pdf')
plt.show()
```
